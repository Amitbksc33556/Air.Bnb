```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("compressed_data.csv")
df
df.head()
```

```python
df.columns
```

```python
# Check missing values
```

```python
print(df.isnull().sum())
```

```python
df.info()
```

```python
df['last review'] = pd.to_datetime(df['last review'], errors = 'coerce')
```

```python
df.info()
```

```python
df.fillna({'reviews per month': 0, 'last review': df['last review'].min()}, inplace=True)
```

```python
df.dropna(subset = ['NAME','host name'],inplace = True)
```

```python
print(df.isnull().sum())
```

```python
df = df.drop(columns = ["license","house_rules"],errors = 'ignore')
```

```python
df.head()
```

```python
# remove $ symbol from the price list
df['price'] = df['price'].replace('[\$,]','',regex = True).astype(float)
df['service fee'] = df['service fee'].replace('[\$,]','',regex = True).astype(float)                                  
```

```python
df.head()
```

```python
# remove duplicate
```

```python
df.drop_duplicates(inplace = True)
df.info()
```

```python
# descriptive statistics
```

```python
df.describe()
```

```python
# Questions,What is the distribution of listing Price
```

```python
plt.figure(figsize = (10,6))
sns.histplot(df['price'],bins = 50,kde = True,color = 'purple')
plt.title("Distribution of listing price")
plt.xlabel("Price $")
plt.ylabel("Frequency")
plt.show()
```

```python
# how are different room types distributed
```

```python
df ['room type']
```

```python
plt.figure(figsize = (8,5))
sns.countplot( x = "room type", data = df,color = "hotpink")
plt.show()
```

```python
# Listings distributed across different neighborhoods
```

```python
plt.figure(figsize = (12,8))
sns.countplot(y = 'neighbourhood group',data = df,color = "lightgreen",order = df['neighbourhood group'].value_counts().index)
plt.title('Number of listings by Neighborhood Group')
plt.xlabel('Count')
plt.ylabel('Neighborhood Group')
plt.show()
```

# what is the relationship between price and room type?


```python
plt.figure(figsize = (10,6))
sns.boxplot(x = 'room type', y = 'price',hue = 'room type',data = df,palette = 'Set1')
plt.title('Price vs Room Type')
plt.xlabel('Room type')
plt.ylabel('Price ($)')
plt.legend(title = 'Room type')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='room type', y='price', data=df)  
plt.title('Price vs Room Type')
plt.xlabel('Room Type')
plt.ylabel('Price ($)')
plt.show()
```

```python
how has the number of reviews change over with time
```

```python
df['last review'] = pd.to_datetime(df['last review'])
reviews_over_time  = df.groupby(df['last review'].dt.to_period('M')).size()

plt.figure(figsize=(12,6))
reviews_over_time.plot(kind= 'line',color = 'red')
plt.title('Number of Reviews over time')
plt.xlabel('Date')
plt.ylabeel('Number of reviews')
plt.show()
```
